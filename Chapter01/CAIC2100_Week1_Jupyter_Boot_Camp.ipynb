## Part 13: Kernel and Notebook Tips

### Restart the kernel when:

- Variables behave unexpectedly.
- You ran cells out of order.
- A library import appears stuck.
- The notebook becomes unresponsive.

In JupyterLab, look under the **Kernel** menu.

### Useful keyboard shortcuts

- `Shift + Enter` — Run the current cell
- `Esc`, then `A` — Add a cell above
- `Esc`, then `B` — Add a cell below
- `Esc`, then `M` — Change a cell to Markdown
- `Esc`, then `Y` — Change a cell to code
- `Esc`, then `D`, `D` — Delete a cell
